```python
count = 0
for i in range(len(pdtb3)):
    scls1b = pdtb3.loc[i, 'SClass2A']
    r = pdtb3.loc[i, 'Relation_Type']
    if type(scls1b) != (np.float64) and type(scls1b) != float and  r == 'Implicit':

        count += 1
```
