```python
class ProcessTree:
    def __init__(self):
        pass
        
    def _simplify_tree(self, tree):
        # tree.label is the word
        # tree.name is arg1 arg2 or pos tag
        temp = []
        if len(tree.children) == 0:
            return tree.label
        for c in tree.children:
            child_label = self._simplify_tree(c)
            temp.append(child_label)
        if len(set(temp)) == 1:
            tree.label = temp[0]
        else:
            tree.label = tree.name
        return tree.label
    
    def simplify_tree(self, tree):
        root_label = self._simplify_tree(tree)
        if tree.label in ['Arg1', 'Arg2', 'Conn', 'none']:
            tree.children = self.get_leave_node(tree)
            return
        for i, c in enumerate(tree.children):
            if self.deeperthan1(c):
                self.simplify_tree(c)
            else:
                n = TreeNode()
                n.children = [c]
                n.label = c.label
                tree.children[i] = n
                
    def more_simplify_tree(self, tree):
        # the next child will be inherited by the first child
        if not self.deeperthan1(tree):
            return 
        tmp = []
        prev_c = None
        for c in tree.children:
            self.more_simplify_tree(c)
            if prev_c == None:
                tmp.append(c)
                prev_c = c
                continue
            if c.name != prev_c.name or c.name not in ['Conn', 'Arg1', 'Arg2', 'none']:
                tmp.append(c)
                prev_c = c
            else:
                prev_c.children = prev_c.children + c.children
        tree.children = tmp

    def deeperthan1(self, c):
        for cc in c.children:
            if len(cc.children) > 0:
                return True
        return False

    def get_leave_node(self, tree):
        leave_nodes = tree.get_leaves()
        preterminals = [n.up for n in leave_nodes]
        return preterminals
    
    def print_tree_label(self, tree):
        """this is for debug purpose and wont be used normally"""
        if self.deeperthan1(tree):
            tree.name = tree.label
        for c in tree.children:
            c = self.print_tree_label(c)
        return tree
    
    def tree2string(self, tree):
        tree_string = "(" + tree.name + " "
        if self.deeperthan1(tree):
            for c in tree.children:
                tree_string += self.tree2string(c)
        else:
            tree_string += tree.children[0].name
        return tree_string + ")"
    
    
    def add_sense(self, tree, sense, toRoot=False):
        if not self.deeperthan1(tree):
            # if the tree is not deeper than 1
            return
        tree.name = sense + '-' + tree.name
        if toRoot:
            return
        else:
            for c in tree.children:
                self.add_sense(c, sense, toRoot)
        return
    
    def get_simplified_tree_string(self, tree, level='normal', sense=None, toRoot=False):
        self.simplify_tree(tree)
        pt.print_tree_label(tree)
        if level == 'normal' and sense!=None:
            self.add_sense(tree, sense, toRoot)
#         print(tree.get_ascii(show_internal=True))
        if level == 'more':
            self.more_simplify_tree(tree)
            if sense != None:
                self.add_sense(tree, sense, toRoot)
#         print(tree.get_ascii(show_internal=True))
        return self.tree2string(tree)
```

```python
ret = []
pt = ProcessTree()
pdtb3.parse_data = pdtb3.parse_data_train
for i in range(len(pdtb3.parse_data)):
    if pdtb3.get_arg_sent_id(i, 'Arg1') == pdtb3.get_arg_sent_id(i, 'Arg2') and len(pdtb3.get_arg_sent_id(i, 'Arg1')) == 1:
        syntax_trees, _ = pdtb3.get_arg1_arg2_parsetree_with_label(i)
        sense = pdtb3.parse_data[i]['Sense'][0].split('.')[0]
        for s in syntax_trees:
#             tree_string, string = pt.get_simplified_tree_string(s.tree)
            tree_string = pt.get_simplified_tree_string(s.tree, 'normal', sense, toRoot=False)
            ret.append(tree_string)
```
