```python
import json
import codecs
import os
```

```python
parse_data_00_path = 'conll-00-01-24/pdtb3-relations.json'
parse_data_00 = [json.loads(o) for o in codecs.open(parse_data_00_path).readlines()]
parse_dict_00_path = 'conll-00-01-24/pdtb-parse.json'
parse_dict_00 = json.loads(codecs.open(parse_dict_00_path).read())

parse_data_rest = [json.loads(o) for o in codecs.open('conll-rest/pdtb3-relations.json')]
parse_dict_rest = json.loads(codecs.open('conll-rest/pdtb-parse.json').read())
```

```python
parse_data_train_path = 'conll-main/train/relations.json'
parse_data_dev_path = 'conll-main/dev/relations.json'
parse_data_test_path = 'conll-main/test/relations.json'
parse_data_train = [json.loads(o) for o in codecs.open(parse_data_train_path).readlines()]
parse_data_dev = [json.loads(o) for o in codecs.open(parse_data_dev_path).readlines()]
parse_data_test = [json.loads(o) for o in codecs.open(parse_data_test_path).readlines()]

parse_dict_train = json.loads(codecs.open('conll-main/train/pdtb-parses.json').read())
parse_dict_dev = json.loads(codecs.open('conll-main/dev/pdtb-parses.json').read())
parse_dict_test = json.loads(codecs.open('conll-main/test/pdtb-parses.json').read())
```

```python
parse_data = parse_data_00 + parse_data_rest + parse_data_train + parse_data_dev + parse_data_test
parse_dict = {}
parse_dict.update(parse_dict_00)
parse_dict.update(parse_dict_rest)
parse_dict.update(parse_dict_train)
parse_dict.update(parse_dict_dev)
parse_dict.update(parse_dict_test)
```

```python
for key in parse_dict.keys():
    with open('conll/parse_dict/' + key, 'w') as file:
        json.dump({key: parse_dict[key]}, file)
```

```python
for key in parse_dict.keys():
    with open('conll/parse_data/' + key, 'w') as file:
        for o in parse_data:
            if o['DocID'] == key:
                file.write(json.dumps(o) + '\n') 
```

## generate dataset

```python
parse_data_train = []
parse_dict_train = {}
for key, value in parse_dict.items():
    if int(key[4:6]) > 1 and int(key[4:6]) < 21:
        print(key)
        parse_dict_train[key] = value
numbers = []
for o in parse_data:
    n = int(o['DocID'][4:6])
    if n > 1 and n < 21:
        numbers.append(n)
        parse_data_train.append(o)
print(sorted(set(numbers)))
with open('conll/train/' + 'pdtb-parses.json', 'w') as f:
    json.dump(parse_dict_train, f)

with open('conll/train/' + 'relations.json', 'w') as f:
    for o in parse_data_train:
        f.write(json.dumps(o) + '\n')
```

```python
parse_data_dev = []
parse_dict_dev = {}
for key, value in parse_dict.items():
    if int(key[4:6]) in [0, 1]:
        print(key)
        parse_dict_dev[key] = value
numbers = []
for o in parse_data:
    n = int(o['DocID'][4:6])
    if n in [0,1]:
        numbers.append(n)
        parse_data_dev.append(o)
print(sorted(set(numbers)))
with open('conll/dev/' + 'pdtb-parses.json', 'w') as f:
    json.dump(parse_dict_dev, f)

with open('conll/dev/' + 'relations.json', 'w') as f:
    for o in parse_data_dev:
        f.write(json.dumps(o) + '\n')
```

```python
parse_data_test = []
parse_dict_test = {}
for key, value in parse_dict.items():
    if int(key[4:6]) in [21, 22]:
        print(key)
        parse_dict_test[key] = value
numbers = []
for o in parse_data:
    n = int(o['DocID'][4:6])
    if n in [21, 22]:
        numbers.append(n)
        parse_data_test.append(o)
print(sorted(set(numbers)))
with open('conll/test/' + 'pdtb-parses.json', 'w') as f:
    json.dump(parse_dict_test, f)

with open('conll/test/' + 'relations.json', 'w') as f:
    for o in parse_data_test:
        f.write(json.dumps(o) + '\n')
```

## generate all

```python
dict_files = os.listdir('conll/all/parse_dict/')
```

```python
parse_dict = {}
for o in dict_files:
    path = 'conll/all/parse_dict/' + o
    parse_dict.update(json.loads(codecs.open(path).read()))
```

```python
data_files = os.listdir('conll/all/parse_data/')
parse_data = []
for o in dict_files:
    path = 'conll/all/parse_data/' + o
    parse_data.extend([json.loads(o) for o in codecs.open(path).readlines()])
```

```python
with open('conll/all/pdtb-parses.json', 'w') as f:
    json.dump(parse_dict, f)
```

```python
with open('conll/all/relations.json', 'w') as f:
    for o in parse_data: f.write(json.dumps(o) + '\n')
```
