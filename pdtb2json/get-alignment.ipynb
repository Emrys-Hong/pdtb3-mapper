```python
import os
from pathlib import Path
import pandas as pd
import re
import json
import codecs
import numpy as np
```

```python
def load_and_merge_parse_dicts(folder):
    folder = Path(folder)
    filenames = os.listdir(folder)
    ret = {}
    for n in filenames:
        ret.update(json.loads(codecs.open(folder/n, encoding='latin-1').read()))
    return ret
```

```python
parse_dict = load_and_merge_parse_dicts('../dataset/pdtb-parses/')
```

```python
words = []
count = []
for docid in parse_dict.keys():
    doc = parse_dict[docid]
    if 'wsj_00' in docid or 'wsj_01' in docid or 'wsj_23' in docid or 'wsj_24' in docid:
        rawtext = open('../dataset/raw/'+docid).read()[7:]
        index = 0
        constant = 40
        for sentence in doc['sentences']:
            for word in sentence['words']:
                word = word[0]
#                 ex = ['``', '0', '*-1', '*-2', '*T*-1', 
#                      '*U*', '-LRB-', '-RRB-', '*ICH*-2']
#                 if word in ex: continue
                if word in rawtext[index: index+len(word)+constant]:
                    word_begin_index = index + rawtext[index: index+len(word)+constant].index(word)
                else:
                    count.append(word)
    #                 print('rawtext',rawtext[index-10: index+len(word)+constant])
    #                 print('word', word)
    #                 continue
                word_end_index = word_begin_index + len(word)
                index = word_end_index
        if index-len(rawtext)<-3:
            print(docid)
```
